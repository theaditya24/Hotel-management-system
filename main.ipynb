# 1. CUSTOMERS SECTION

### 1.1 : Add Customers

```python
import csv

def generate_customer_id():   
    with open("customers.csv", "r") as file:
        reader = csv.reader(file)
        rows = list(reader)

        if len(rows) <= 1:
            return 1
        
        last_row = rows[-1]
        last_id = int(last_row[0])
        
        return last_id + 1
```

```python
def add_customer():
    print("------ ADD DETAILS OF CUSTOMER -----")
    cust_name = input("Enter name of customer : ")
    cust_email = input("Enter email of customer : ")
    cust_phone = input("Enter phone number(10-DIGIT) of customer : ")
    cust_address = input("Enter the address of customer : ")
    cust_prooftype = input("Enter Type of Proof of identity of customer : ")
    cust_id = generate_customer_id()
    cust_status = input("Enter the status of the customer (Active/Inactive) : ")
    customer = [cust_id,cust_name,cust_phone,cust_email,cust_address,cust_prooftype,cust_status]

    with open("customers.csv","a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(customer)
```

### 1.2 : VIEW CUSTOMERS

```python
def view_customers():
    print("------ CUSTOMER LIST ------")
    
    with open("customers.csv","r") as file:
        reader = csv.reader(file)
        
        for row in reader:
            print(row)
```

### 1.3 : Search Customer by ID / Name / Phone

```python
import csv

def search_customer():
    print("------ SEARCH CUSTOMER ------")
    print("1. Search by ID")
    print("2. Search by Name")
    print("3. Search by Phone")

    choice = input("Enter your choice: ")
    value = input("Enter search value: ")

    found = False

    with open("customers.csv", "r") as file:
        reader = csv.reader(file)
        next(reader)

        for row in reader:
            if choice == "1" and row[0] == value:
                print_customer(row)
                found = True

            elif choice == "2" and row[1].lower() == value.lower():
                print_customer(row)
                found = True

            elif choice == "3" and row[2] == value:
                print_customer(row)
                found = True

    if not found:
        print("Customer not found.")
```

# 2. ROOM MANAGEMENT

### 2.1 : ADD ROOM

```python
import csv

def generate_room_id():
    with open("rooms.csv","r") as file:
        reader = csv.reader(file)
        rows = list(reader)

        if len(rows) <= 1:
            return 1

        last_row = rows[-1]
        last_id = int(last_row[0])

        return last_id + 1
```

```python
def add_room():
    print("------ ADD ROOM ------")

    room_type = input("Enter room type (Single/Double/Deluxe): ")
    price = input("Enter price per night: ")
    status = "Available"

    room_id = generate_room_id()

    room = [room_id, room_type, price, status]

    with open("rooms.csv","a",newline="") as file:
        writer = csv.writer(file)
        writer.writerow(room)

    print("Room added successfully.")
```

### 2.2 : View All Rooms

```python
def view_rooms():
    print("------ ROOM LIST ------")

    with open("rooms.csv","r") as file:
        reader = csv.reader(file)

        for row in reader:
            print(row)
```

### 2.3 : View Available Rooms

```python
def view_available_rooms():
    print("------ AVAILABLE ROOMS ------")

    with open("rooms.csv","r") as file:
        reader = csv.reader(file)
        next(reader)

        for row in reader:
            if row[3] == "Available":
                print(row)
```

### 2.4 : Update Room Status

```python
def update_room_status():
    room_id = input("Enter room ID to update status: ")

    rows = []

    with open("rooms.csv","r") as file:
        reader = csv.reader(file)

        for row in reader:
            if row[0] == room_id:
                new_status = input("Enter new status (Available/Booked/Maintenance): ")
                row[3] = new_status

            rows.append(row)

    with open("rooms.csv","w",newline="") as file:
        writer = csv.writer(file)
        writer.writerows(rows)

    print("Room status updated successfully.")
```

### 2.5 : Room Management Menu

```python
def room_menu():
    while True:
        print("\n------ ROOM MANAGEMENT ------")
        print("1. Add Room")
        print("2. View Rooms")
        print("3. View Available Rooms")
        print("4. Update Room Status")
        print("5. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add_room()

        elif choice == "2":
            view_rooms()

        elif choice == "3":
            view_available_rooms()

        elif choice == "4":
            update_room_status()

        elif choice == "5":
            break

        else:
            print("Invalid choice")
```

# 3. BOOKING MANAGEMENT

```python
import csv
from datetime import datetime

def generate_booking_id():
    with open("bookings.csv","r") as file:
        reader = csv.reader(file)
        rows = list(reader)

        if len(rows) <= 1:
            return 1

        last_row = rows[-1]
        last_id = int(last_row[0])

        return last_id + 1
```

### 3.1 : Check if customer exists

```python
def check_customer_exists(customer_id):

    with open("customers.csv","r") as file:
        reader = csv.reader(file)
        next(reader)

        for row in reader:
            if row[0] == customer_id and row[6] == "Active":
                return True

    return False
```

### 3.2 : Show Available Rooms

```python
def show_available_rooms():

    print("------ AVAILABLE ROOMS ------")

    with open("rooms.csv","r") as file:
        reader = csv.reader(file)
        next(reader)

        for row in reader:
            if row[3] == "Available":
                print("Room ID:",row[0],"Type:",row[1],"Price:",row[2])
```

### 3.3 : Get Room Price

```python
def get_room_price(room_id):

    with open("rooms.csv","r") as file:
        reader = csv.reader(file)
        next(reader)

        for row in reader:
            if row[0] == room_id:
                return int(row[2])
```

### 3.4 : Update Room Status

```python
def update_room_status(room_id,new_status):

    rows = []

    with open("rooms.csv","r") as file:
        reader = csv.reader(file)

        for row in reader:
            if row[0] == room_id:
                row[3] = new_status
            rows.append(row)

    with open("rooms.csv","w",newline="") as file:
        writer = csv.writer(file)
        writer.writerows(rows)
```

```python
### 3.5 : Book Room (Main Function)
```

```python
def book_room():

    print("------ ROOM BOOKING ------")

    customer_id = input("Enter customer ID: ")

    if not check_customer_exists(customer_id):
        print("Customer not found or inactive")
        return

    show_available_rooms()

    room_id = input("Enter room ID: ")

    check_in = input("Enter check-in date (YYYY-MM-DD): ")
    check_out = input("Enter check-out date (YYYY-MM-DD): ")

    checkin_date = datetime.strptime(check_in,"%Y-%m-%d")
    checkout_date = datetime.strptime(check_out,"%Y-%m-%d")

    nights = (checkout_date - checkin_date).days

    price = get_room_price(room_id)

    total_amount = nights * price

    booking_id = generate_booking_id()

    status = "Active"

    booking = [booking_id,customer_id,room_id,check_in,check_out,total_amount,status]

    with open("bookings.csv","a",newline="") as file:
        writer = csv.writer(file)
        writer.writerow(booking)

    update_room_status(room_id,"Booked")

    print("Room booked successfully")
    print("Total amount:",total_amount)
```

### 3.5 : Checkout Logic

```python
def checkout_customer():

    booking_id = input("Enter booking ID: ")

    rows = []
    room_id = None

    with open("bookings.csv","r") as file:
        reader = csv.reader(file)

        for row in reader:
            if row[0] == booking_id:
                row[6] = "Completed"
                room_id = row[2]
            rows.append(row)

    with open("bookings.csv","w",newline="") as file:
        writer = csv.writer(file)
        writer.writerows(rows)

    if room_id:
        update_room_status(room_id,"Available")
        print("Checkout successful")
    else:
        print("Booking not found")
```

### 3.6 : Cancel Booking

```python
def cancel_booking():

    booking_id = input("Enter booking ID to cancel: ")

    rows = []
    room_id = None
    found = False

    with open("bookings.csv","r") as file:
        reader = csv.reader(file)

        for row in reader:
            if row[0] == booking_id and row[6] == "Active":
                row[6] = "Cancelled"
                room_id = row[2]
                found = True
            rows.append(row)

    with open("bookings.csv","w",newline="") as file:
        writer = csv.writer(file)
        writer.writerows(rows)

    if found:
        update_room_status(room_id,"Available")
        print("Booking cancelled successfully.")
    else:
        print("Booking not found or already completed/cancelled.")
```

### 3.6 : Booking Menu

```python
def booking_menu():

    while True:

        print("\n------ BOOKING MANAGEMENT ------")
        print("1 Book Room")
        print("2 Checkout Customer")
        print("3 Cancel Booking")
        print("4 Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            book_room()

        elif choice == "2":
            checkout_customer()

        elif choice == "3":
            cancel_booking()

        elif choice == "4":
            break

        else:
            print("Invalid choice")
```

# 4. PAYMENT MANAGEMENT

### 4.1 : Generate Payment ID

```python
import csv
from datetime import datetime

def generate_payment_id():

    with open("payments.csv","r") as file:
        reader = csv.reader(file)
        rows = list(reader)

        if len(rows) <= 1:
            return 1

        last_row = rows[-1]
        last_id = int(last_row[0])

        return last_id + 1
```

### 4.2 : Get Booking Amount

```python
def get_booking_amount(booking_id):

    with open("bookings.csv","r") as file:
        reader = csv.reader(file)
        next(reader)

        for row in reader:
            if row[0] == booking_id and row[6] == "Active":
                return row[5]

    return None
```

### 4.3 : Make Payment

```python
def make_payment():

    print("------ PAYMENT ------")

    booking_id = input("Enter booking ID: ")

    amount = get_booking_amount(booking_id)

    if amount is None:
        print("Booking not found or already completed.")
        return

    print("Total amount to pay:", amount)

    payment_method = input("Enter payment method (Cash/Card/UPI): ")

    payment_id = generate_payment_id()

    payment_date = datetime.today().strftime("%Y-%m-%d")

    status = "Paid"

    payment = [payment_id,booking_id,amount,payment_date,payment_method,status]

    with open("payments.csv","a",newline="") as file:
        writer = csv.writer(file)
        writer.writerow(payment)

    print("Payment successful.")
```

### 4.4 : View Payments

```python
def view_payments():

    print("------ PAYMENT LIST ------")

    with open("payments.csv","r") as file:
        reader = csv.reader(file)

        for row in reader:
            print(row)
```

### 4.5 : Payment Menu

```python
def payment_menu():

    while True:

        print("\n------ PAYMENT MODULE ------")
        print("1 Make Payment")
        print("2 View Payments")
        print("3 Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            make_payment()

        elif choice == "2":
            view_payments()

        elif choice == "3":
            break

        else:
            print("Invalid choice")
```
